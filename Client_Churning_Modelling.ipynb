### Data Preprocessing and Feature Engineering

```python
# creating a copy of the original data
data1 = data.copy()
```

```python
# dropping the id column
data1.drop("id", axis = 1, inplace = True)
```

```python
# replacing 1 for "t" and 0 for "f" in the has_gas column
data1["has_gas"] = data1["has_gas"].replace({"t": 1, "f": 0})
```

```python
# converting the price_date column into month_price
data1["price_date"] = pd.to_datetime(data1["price_date"])

data1["month"] = data1["price_date"].dt.month
```

```python
# dropping the price_date column
data1.drop("price_date", axis = 1, inplace = True)
```
