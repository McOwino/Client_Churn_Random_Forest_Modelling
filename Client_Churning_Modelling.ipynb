### Outlier Removal

#### Visualizing the effect of outliers using boxplot

```python
for column in data1.drop(['channel_sales','has_gas', 'month','origin_up', 'churn'], axis = 1).columns:
    plt.figure(figsize = (10,8))
    sns.boxplot(x= column , color = "r", data = data1)
    plt.title(f"The Boxplot of the {column}")
    plt.show()
```

```python
def outlier_remover(data):
    data_new = data.copy()
    
    for column in data.drop(['channel_sales', 'has_gas', 'month', 'origin_up', 'churn'], axis=1).columns:
        Q1 = data_new[column].quantile(0.25)
        Q3 = data_new[column].quantile(0.75)
        IQR = Q3 - Q1
        multiplier = 1.5
        LF = Q1 - (IQR * multiplier)
        UF = Q3 + (IQR * multiplier)
        outliers = data_new[(data_new[column] > UF) | (data_new[column] < LF)]
        data_new = data_new[~data_new.index.isin(outliers.index)]
    
    return data_new
```

```python
data2 =  outlier_remover(data1)

data2.info()
```

```python
#visualizing the effect of teh above operation using boxplot
for column in data2.drop(['channel_sales','has_gas', 'month','origin_up', 'churn'], axis = 1).columns:
    plt.figure(figsize = (10,8))
    sns.boxplot(x= column , color = "purple", data = data2)
    plt.title(f"The Boxplot of the {column}")
    plt.show()
```

### TrainIng, Testing and Splitting

```python
# Modelling Libraries
from sklearn.model_selection import train_test_split,cross_val_score,ShuffleSplit
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
# separating teh target variables from the rest of the features
X = data2.drop("churn", axis = 1)

y = data2["churn"]

# scalling the X and Y variables using minmaxscaller
scaler=MinMaxScaler()
scaler.fit(X)
X=scaler.fit_transform(X)
```

```python
print(f"The shape of X is {X.shape}\n")

print(f"The shape of y is {y.shape}\n")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

print(f"The shape of X_train is {X_train.shape}\n")
print(f"The shape of X_test is {X_test.shape}\n")
print(f"The shape of y_train is {y_train.shape}\n")
print(f"The shape of y_test is {y_test.shape}\n")
```

#### Class Balancing

```python
# Balancing the class weights with SMOTETomek Procedure
 # importing the libraries and models for class balancing
from imblearn.combine import SMOTETomek
# Instantiate SMOTETomek
smote_tomek = SMOTETomek(random_state=42)

# Apply SMOTETomek to the training data
X_resampled, y_resampled = smote_tomek.fit_resample(X_train, y_train)

print(f"The shape of X_resampled is {X_resampled.shape}\n")

print(f"The shape of y_resampled is {y_resampled.shape}\n")
```

### Random Forest Classifier

```python
# initializing the random forest model
rf = RandomForestClassifier()

# training the model on the X_resampled and y_resampled
rf.fit(X_resampled, y_resampled)

# evaluating the perfomance of the model on the trained data
rf.score(X_resampled, y_resampled)
```
